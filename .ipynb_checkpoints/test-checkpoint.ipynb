```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.preprocessing import LabelEncoder
import joblib
import os
```

```python
# Definir o número máximo de linhas e colunas a serem exibidos
pd.set_option('display.max_rows', None)  # Exibe todas as linhas
pd.set_option('display.max_columns', None)  # Exibe todas as colunas
```

```python
df = pd.read_csv("dados/test.csv")
df.head(10)
```

```python
len(df)
```

```python
# Calcular a porcentagem de valores nulos em cada coluna
percent_null = (df.isnull().sum() / len(df)) * 100

# Filtrar e exibir apenas as colunas com porcentagem de nulos maior que 0
percent_null = percent_null[percent_null > 0]

percent_null
```

```python
df['LotFrontage'] = df['LotFrontage'].fillna(0)
df['MasVnrArea'] = df['MasVnrArea'].fillna(0)
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(0)
```

```python
# Dicionário de mapeamento
quality_mapping = {
    'Ex': 5,
    'Gd': 4,
    'TA': 3,
    'Fa': 2,
    'Po': 1,
    'NA': 0  # Substituindo NA por 0 ou outro valor desejado
}

# Colunas a serem convertidas
quality_columns = [
    'ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond',
    'HeatingQC', 'KitchenQual', 'FireplaceQu', 
    'GarageQual', 'GarageCond', 'PoolQC'
]

# Aplica a substituição em cada coluna
for col in quality_columns:
    df[col] = df[col].map(quality_mapping)
```

```python
df[quality_columns] = df[quality_columns].fillna(0)
```

```python
df.head(10)
```

```python
# Identifica as colunas categóricas (não numéricas)
categorical_cols = df.select_dtypes(include=['object']).columns

# Aplica o Label Encoding ao conjunto de teste
for col in categorical_cols:
    # Inicializa um novo encoder
    new_encoder = LabelEncoder()

    try:
        # Tenta carregar o encoder existente
        encoder = joblib.load(f'label_encoder/{col}_label_encoder.joblib')
        
        # Verifica valores que não estão no encoder
        unseen_values = df[col][~df[col].isin(encoder.classes_)].unique()
        
        if len(unseen_values) > 0:
            print(f"Valores não vistos encontrados na coluna '{col}': {unseen_values}. Criando um novo encoder.")
            
            # Obtém os valores únicos da coluna e adiciona os valores não vistos
            unique_values = encoder.classes_.tolist() + unseen_values.tolist()
            
            # Ajusta o novo encoder
            new_encoder.fit(unique_values)
            
            # Aplica a transformação
            df[col] = new_encoder.transform(df[col])
            
            # Salva o novo encoder
            joblib.dump(new_encoder, f'label_encoder/{col}_label_encoder.joblib')
            print(f"Novo encoder para a coluna '{col}' criado e salvo.")
        else:
            # Aplica a transformação se não houver valores não vistos
            df[col] = encoder.transform(df[col])
    
    except FileNotFoundError:
        print(f"Encoder para a coluna '{col}' não encontrado. Criando um novo.")
        
        # Obtenha todos os valores únicos da coluna
        unique_values = df[col].unique().tolist()
        
        # Ajusta o novo encoder
        new_encoder.fit(unique_values)
        
        # Aplica a transformação
        df[col] = new_encoder.transform(df[col])
        
        # Salva o novo encoder
        joblib.dump(new_encoder, f'label_encoder/{col}_label_encoder.joblib')
        print(f"Novo encoder para a coluna '{col}' criado e salvo.")
    
    except ValueError as e:
        print(f"Erro ao transformar a coluna '{col}': {e}")

        # Obtenha todos os valores únicos da coluna
        unique_values = df[col].unique().tolist()
        
        # Ajusta o novo encoder
        new_encoder.fit(unique_values)
        
        # Aplica a transformação
        df[col] = new_encoder.transform(df[col])
        
        # Salva o novo encoder
        joblib.dump(new_encoder, f'label_encoder/{col}_label_encoder.joblib')
        print(f"Novo encoder para a coluna '{col}' criado e salvo.")
```

```python
# Criação de novas colunas através da engenharia de recursos

# 1. Idade da Casa
df['HouseAge'] = df['YrSold'] - df['YearBuilt']

# 2. Idade da Reforma
df['RemodelAge'] = df['YrSold'] - df['YearRemodAdd']

# 3. Área Total
df['TotalArea'] = df['GrLivArea'] + df['TotalBsmtSF'] + df['GarageArea']

# 4. Número Total de Banheiros
df['TotalBathrooms'] = df['FullBath'] + df['HalfBath'] + df['BsmtFullBath'] + df['BsmtHalfBath']

# 5. Proporção de Quartos para Banheiros
df['BedroomBathroomRatio'] = df['BedroomAbvGr'] / df['TotalBathrooms'].replace(0, np.nan)  # Evita divisão por zero

# 6. Qualidade Geral do Acabamento
df['OverallFinishQuality'] = df['OverallQual'] + df['OverallCond']  # Pode ser ponderado se desejado

# 7. Área de Decks e Varandas
df['TotalOutdoorArea'] = (df['WoodDeckSF'] + df['OpenPorchSF'] + 
                          df['EnclosedPorch'] + df['3SsnPorch'] + 
                          df['ScreenPorch'])

# 8. Tamanho do Lote por Quarto
df['LotSizePerRoom'] = df['LotArea'] / df['BedroomAbvGr'].replace(0, np.nan)  # Evita divisão por zero

# 11. Qualidade da Área de Garagem
df['GarageQuality'] = df['GarageQual'] + df['GarageCond']  # Ajuste conforme necessário

# 12. Proporção de Área de Garagem em Relação ao Lote
df['GarageToLotRatio'] = df['GarageArea'] / df['LotArea'].replace(0, np.nan)  # Evita divisão por zero

# 13. Qualidade do Acabamento Externo
df['ExteriorQuality'] = df['ExterQual'] + df['ExterCond']  # Pode ser ponderado se desejado

# 14. Qualidade do Porão
df['BasementQuality'] = df['BsmtQual'] + df['BsmtCond']

# 15. Qualidade da Garagem
df['GarageQuality'] = df['GarageQual'] + df['GarageCond']
```

```python
# Salvar o DataFrame em um arquivo CSV
df.to_csv('dados/test_dados_processados.csv', index=False)
```
