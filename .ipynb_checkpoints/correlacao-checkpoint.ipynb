```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from scipy.stats import chi2_contingency
from sklearn.preprocessing import LabelEncoder
import joblib
```

```python
# Definir o número máximo de linhas e colunas a serem exibidos
pd.set_option('display.max_rows', None)  # Exibe todas as linhas
pd.set_option('display.max_columns', None)  # Exibe todas as colunas
```

```python
df = pd.read_csv("dados/train_dados_processados.csv")
df.head(10)
```

```python
# Inicializar a lista resultados
resultados = []

# Calculando a correlação de Pearson entre as variáveis independentes e a variável alvo
correlacao = df.corr(method='pearson')

# Exibindo a correlação com a variável alvo 'VIVO'
correlacao_vivo = correlacao['SalePrice'].sort_values(ascending=False)

# Exibindo as correlações
print(correlacao_vivo*100)
resultados.append(correlacao_vivo)

# Gerando o gráfico de barras horizontais
plt.figure(figsize=(10, 8))
correlacao_vivo.drop('SalePrice').plot(kind='barh', color='skyblue')  # Excluindo 'VIVO' de si mesma
plt.title('Correlação de Pearson com a variável SalePrice')
plt.xlabel('Correlação')
plt.ylabel('Variáveis')
plt.grid(axis='x')

# Exibir o gráfico
plt.show()
```

```python
# Inicializar a lista resultados
resultados_spearman = []

# Calculando a correlação de Spearman entre as variáveis independentes e a variável alvo
correlacao_spearman = df.corr(method='spearman')

# Exibindo a correlação com a variável alvo 'VIVO'
correlacao_vivo_spearman = correlacao_spearman['SalePrice'].sort_values(ascending=False)

# Exibindo as correlações de Spearman
print(correlacao_vivo_spearman*100)
resultados_spearman.append(correlacao_vivo_spearman)

# Gerando o gráfico de barras horizontais - Spearman
plt.figure(figsize=(10, 8))
correlacao_vivo_spearman.drop('SalePrice').plot(kind='barh', color='orange')  # Excluindo 'VIVO' de si mesma
plt.title('Correlação de Spearman com a variável SalePrice')
plt.xlabel('Correlação')
plt.ylabel('Variáveis')
plt.grid(axis='x')

# Exibir o gráfico
plt.show()
```
