Remover nulos -> Electrical

```python
# Remover colunas específicas do DataFrame
df = df.drop(columns='Id')
```

```python
df = df.dropna(subset=['Electrical'])
```

```python
df['LotFrontage'] = df['LotFrontage'].fillna(0)
df['MasVnrArea'] = df['MasVnrArea'].fillna(0)
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(0)
```

```python
# Dicionário de mapeamento
quality_mapping = {
    'Ex': 5,
    'Gd': 4,
    'TA': 3,
    'Fa': 2,
    'Po': 1,
    'NA': 0  # Substituindo NA por 0 ou outro valor desejado
}

# Colunas a serem convertidas
quality_columns = [
    'ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond',
    'HeatingQC', 'KitchenQual', 'FireplaceQu', 
    'GarageQual', 'GarageCond', 'PoolQC'
]

# Aplica a substituição em cada coluna
for col in quality_columns:
    df[col] = df[col].map(quality_mapping)
```

```python
df[quality_columns] = df[quality_columns].fillna(0)
```

```python
# Identifica as colunas categóricas (não numéricas)
categorical_cols = df.select_dtypes(include=['object']).columns

# Dicionário para armazenar os encoders
label_encoders = {}

# Aplica o Label Encoding a cada coluna categórica
for col in categorical_cols:
    labelencoder = LabelEncoder()
    df[col] = labelencoder.fit_transform(df[col])
    label_encoders[col] = labelencoder  # Salva o encoder no dicionário

# Salva os encoders em arquivos
for col, encoder in label_encoders.items():
    joblib.dump(encoder, f'label_encoder/{col}_label_encoder.joblib')
```

```python
# Criação de novas colunas através da engenharia de recursos

# 1. Idade da Casa
df['HouseAge'] = df['YrSold'] - df['YearBuilt']

# 2. Idade da Reforma
df['RemodelAge'] = df['YrSold'] - df['YearRemodAdd']

# 3. Área Total
df['TotalArea'] = df['GrLivArea'] + df['TotalBsmtSF'] + df['GarageArea']

# 4. Número Total de Banheiros
df['TotalBathrooms'] = df['FullBath'] + df['HalfBath'] + df['BsmtFullBath'] + df['BsmtHalfBath']

# 5. Proporção de Quartos para Banheiros
df['BedroomBathroomRatio'] = df['BedroomAbvGr'] / df['TotalBathrooms'].replace(0, np.nan)  # Evita divisão por zero

# 6. Qualidade Geral do Acabamento
df['OverallFinishQuality'] = df['OverallQual'] + df['OverallCond']  # Pode ser ponderado se desejado

# 7. Área de Decks e Varandas
df['TotalOutdoorArea'] = (df['WoodDeckSF'] + df['OpenPorchSF'] + 
                          df['EnclosedPorch'] + df['3SsnPorch'] + 
                          df['ScreenPorch'])

# 8. Tamanho do Lote por Quarto
df['LotSizePerRoom'] = df['LotArea'] / df['BedroomAbvGr'].replace(0, np.nan)  # Evita divisão por zero

# 11. Qualidade da Área de Garagem
df['GarageQuality'] = df['GarageQual'] + df['GarageCond']  # Ajuste conforme necessário

# 12. Proporção de Área de Garagem em Relação ao Lote
df['GarageToLotRatio'] = df['GarageArea'] / df['LotArea'].replace(0, np.nan)  # Evita divisão por zero

# 13. Qualidade do Acabamento Externo
df['ExteriorQuality'] = df['ExterQual'] + df['ExterCond']  # Pode ser ponderado se desejado

# 14. Qualidade do Porão
df['BasementQuality'] = df['BsmtQual'] + df['BsmtCond']

# 15. Qualidade da Garagem
df['GarageQuality'] = df['GarageQual'] + df['GarageCond']
```

```python
# Salvar o DataFrame em um arquivo CSV
df.to_csv('dados/train_dados_processados.csv', index=False)
```
